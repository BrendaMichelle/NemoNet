```python
import os
import pandas as pd
```

```python
'''
Each row contains the catch_id that links to the images.
Also contained in those rows are the user-defined species_ids
as well as the longitude and latitude of the catch.
'''
catchData = pd.read_csv("../fishdata/Fishackathon_CatchData.csv")
'''
Row for each catch_id.
Each row contains the catch_id that links to the images.
Also contained in those rows are the user-defined species_ids as well as the longitude and latitude of the catch.
'''
speciesData = pd.read_csv("../fishdata/Fishackathon_SpeciesData.csv")
```

```python
## Initialize species id to species name map
idToSpecies = {}
# s_id => s_name
for index, row in speciesData.iterrows():
    idToSpecies[row["s_id"]] = row["s_name"]
```

```python
## Create directories for each species if they don't exist
PhotosBySpeciesDir = "../fishdata/PhotosBySpecies"
for speciesName in idToSpecies.values():
    os.mkdir(PhotosBySpeciesDir + "/" + speciesName)
```

```python
'''
Associate photos with species tag
'''
'''
Each image within that file has a filename that consists of ####_#######_#.png.
The first set of numbers is the most important to you.
Those are the numbers that will link that image to the data contained in the .CSV file
Those first numbers are the catch_id numbers and are the unique identifiers for each catch logged in NetFish.
'''
AllPhotosPath = "../fishdata/AllPhotos"
PhotosBySpeciesDir = "../fishdata/PhotosBySpecies"

speciesDict = {}
for subdir, dirs, files in os.walk(AllPhotosPath):
    for file in files:
        if (file[-4:] == ".jpg"):
            # translate file's catch_id to species_id
            catch_id = file.split("_")[0]
            species_row = catchData.loc[catchData["catch_id"] == int(catch_id)]
            #print(species_row)
            species_id = (species_row["catch_s_id"])
            species_id = species_id.asobject
            if (len(species_id) == 0):
                continue
            species_id = species_id[0]
            #print(idToSpecies[species_id])
            #print(PhotosBySpeciesDir + "/" + idToSpecies[species_id] + "/" + file)
            if (species_id in idToSpecies):
                #print(PhotosBySpeciesDir + "/" + idToSpecies[species_id] + "/" + file)
                os.rename(subdir + "/" + file, PhotosBySpeciesDir + "/" + idToSpecies[species_id] + "/" + file)
```

```python
count = 0
for subdir, dirs, files in os.walk(PhotosBySpeciesDir):
    count += (len(files))
print(count)
```

```python
# Sanity check, get counts for each species
speciesCounts = {}

PhotosBySpeciesDir = "../fishdata/PhotosBySpecies"
for val in idToSpecies.values():
    speciesCounts[val] = 0
for subdir, dirs, files in os.walk(PhotosBySpeciesDir):
    for file in files:
        if (file[-4:] == ".jpg"):
            catch_id = file.split("_")[0]
            species_row = catchData.loc[catchData["catch_id"] == int(catch_id)]
            #print(species_row)
            species_id = (species_row["catch_s_id"])
            species_id = species_id.asobject
            if (len(species_id) == 0):
                continue
            species_id = species_id[0]
            speciesCounts[idToSpecies[species_id]] += 1
            
for key in speciesCounts:
    print("{} : {}".format(key, speciesCounts[key]))
```
